```python
import zipfile
import os

# Path to the uploaded zip file
zip_file_path = 'archive.zip'
extracted_folder_path = 'extracted_data/'

# Extract the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extracted_folder_path)

# List the extracted files and folders
extracted_files = os.listdir(extracted_folder_path)
extracted_files
```

```python
import pandas as pd

# Load the datasets
training_data_path = os.path.join(extracted_folder_path, 'twitter_training.csv')
validation_data_path = os.path.join(extracted_folder_path, 'twitter_validation.csv')

# Read the CSV files
training_df = pd.read_csv(training_data_path)
validation_df = pd.read_csv(validation_data_path)

# Display the first few rows of each dataset
training_head = training_df.head()
validation_head = validation_df.head()

training_head, validation_head
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Renaming columns for clarity
training_df.columns = ['ID', 'Brand', 'Sentiment', 'Text']
validation_df.columns = ['ID', 'Brand', 'Sentiment', 'Text']

# Plotting the sentiment distribution in the training data
plt.figure(figsize=(14, 8))
sns.countplot(data=training_df, x='Brand', hue='Sentiment', palette='viridis')
plt.title('Sentiment Distribution by Brand/Topic in Training Data')
plt.xlabel('Brand/Topic')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Sentiment')
plt.show()
```
